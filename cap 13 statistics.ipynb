```python
from sklearn.datasets import make_blobs
```

```python
X,y=make_blobs(n_samples=250,centers=4,random_state=500,cluster_std=1.25)
```

```python
X
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],s=50)
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=4,random_state=0)
```

```python
model.fit(X)
```

```python
y_kmeans=model.predict(X)
```

```python
y_kmeans[:12]
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y_kmeans,cmap='coolwarm');
```

```python
from sklearn.datasets import make_classification
```

```python
n_samples=100
```

```python
X,y=make_classification(n_samples=n_samples,n_features=2,n_informative=2,n_redundant=0,n_repeated=0,random_state=250)
```

```python
X[:5]
```

```python
X.shape
```

```python
y[:5]
```

```python
y.shape
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=X[:,0],y=X[:,1],c=y,cmap='coolwarm')
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
model=GaussianNB()
```

```python
model.fit(X,y)
```

```python
model.predict_proba(X).round(4)[:5]
```

```python
pred=model.predict(X)
```

```python
pred
```

```python
pred==y
```

```python
accuracy_score(y,pred)
```

```python
Xc=X[y==pred]
Xf=X[y!=pred]
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=Xc[:,0],y=Xc[:,1],c=y[y==pred],marker='o',cmap='coolwarm')
plt.scatter(x=Xf[:,0],y=Xf[:,1],c=y[y!=pred],marker='x',cmap='coolwarm')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression(C=1,solver='lbfgs')
```

```python
model.fit(X,y)
```

```python
model.predict_proba(X).round(4)[:5]
```

```python
pred=model.predict(X)
```

```python
pred
```

```python
accuracy_score(y,pred)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y = train_test_split(X,y,test_size=0.33,random_state=0)
```

```python
model=SVC(C=1,kernel='linear')
```

```python
model.fit(train_x,train_y)
```

```python
pred_train=model.predict(train_x)
```

```python
accuracy_score(train_y,pred_train)
```

```python
pred_test=model.predict(test_x)
```

```python
test_y==pred_test
```

```python
accuracy_score(test_y,pred_test)
```

```python
test_c=test_x[test_y == pred_test]
test_f=test_x[test_y != pred_test]
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=test_c[:,0],y=test_c[:,1],c=test_y[test_y == pred_test],marker='o',cmap='coolwarm')
plt.scatter(x=test_f[:,0],y=test_f[:,0],c=test_y[test_y != pred_test],marker='x',cmap='coolwarm')
```
